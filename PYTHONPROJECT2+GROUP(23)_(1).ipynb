```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("candidates.xlsx")
```

```python
#1.Remove the NaN coloumns. Remove the Rows that contain NaN. Drop all the irrelevnt coloumns.

df=df.dropna()
df
```

```python
df=df.drop(['Username'],axis=1)
df
```

```python
#2.Rename the coloumn names to proper ones.
df = df.rename(columns={'Email': 'Email Address', 'Name': 'Full Name', 'Alloted Set': 'Set'})
df
```

```python
#3.Capitalize the first letter of the names of students
df['Full Name']=df['Full Name'].str.capitalize()
df
```

```python
#15.Find out if there is any student who is topper in both above questions (section wise). 
#Such students will be called scholars. Prints the names of all scholar
scholars=[]
for i in range(len(a1['Full Name'])):
    for j in range(len(topperslist)):
        if(a1.iloc[i,1]==topperslist[j]):
            scholars.append(a1.iloc[i,1])

for i in range(len(a2['Full Name'])):
    for j in range(len(topperslist)):
        if(a2.iloc[i,1]==topperslist[j]):
            scholars.append(a2.iloc[i,1])
            
for i in range(len(a3['Full Name'])):
    for j in range(len(topperslist)):
        if(a3.iloc[i,1]==topperslist[j]):
            scholars.append(a3.iloc[i,1])
            
for i in range(len(a4['Full Name'])):
    for j in range(len(topperslist)):
        if(a4.iloc[i,1]==topperslist[j]):
            scholars.append(a4.iloc[i,1])
            
for i in range(len(a5['Full Name'])):
    for j in range(len(topperslist)):
        if(a5.iloc[i,1]==topperslist[j]):
            scholars.append(a5.iloc[i,1])
            
print("The scholars are:\nNAME:\t{}".format(scholars))
```

```python
#17.Find the count of all students whose Incorrect/Correct Ratio in all sections is more than 1.
df['icr1']=df['Wrong Answers 1']/df['Marks 1']
df['icr2']=df['Wrong Answers 2']/df['Marks 2']
df['icr3']=df['Wrong Answers 3']/df['Marks 3']
df['icr4']=df['Wrong Answers 4']/df['Marks 4']
df['icr5']=df['Wrong Answers 5']/df['Marks 5']
count=0
count1=0
for i in range(df['Full Name'].count()):
    try:
        if ((df.loc[i,'icr1']>1 and df.loc[i,'Marks 1']!=0 ) and (df.loc[i,'icr2']>1 and df.loc[i,'Marks 2']!=0) and (df.loc[i,'icr3']>1 and df.loc[i,'Marks 3']!=0) and (df.loc[i,'icr4']>1 and df.loc[i,'Marks 4']!=0 ) and (df.loc[i,'icr5']>1 and df.loc[i,'Marks 5']!=0)):
            count+=1
    except:
        count1+=1
print('Students whose Incorrect/Correct Ratio in all sections is more than 1:',count)
print(count1)
```

```python
#18.BONUS QUESTION: For the scholars mentioned above, extract their test result from the original data and convert it in a table of marks format. Write it to a text file. 
#Then write a code to send the mail to all those candidates congratulating them for good performance. (2 points)


import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

for i in range(len(scholars)):
    
    scholardf=df[df['Full Name']==scholars[i]]
    fromaddr="atharvapython@gmail.com"
    toaddr=scholardf.iloc[i,2]
    msg=MIMEMultipart()
    msg['From']=fromaddr
    msg["To"]=str(toaddr)
    msg['Subject']=""
    body=""
    f1=open("C:\\Users\\ATHARVA\\Desktop\\passingletter.txt","w")
    f1.write("Best wishes for passing the exam and good luck for more excellent achievement in the future. Congratulations. \
             \nName of the Student:{}\nTotal marks scored:{}".format(scholardf.iloc[i,1],scholardf.iloc[i,4]))
    f1.close()
    msg.attach(MIMEText(body,'plain'))
    filename="passingletter.txt"
    attachment=open("C:\\Users\\ATHARVA\\Desktop\\passingletter.txt")

    part=MIMEBase('application','octet-stream')
    part.set_payload((attachment).read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition',"attachment; filename=%s" %(filename))
    msg.attach(part)


    server=smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login(fromaddr,"pythonatharva")
    text=msg.as_string()
    server.sendmail(fromaddr,toaddr,text)
    print("Email to {} sent succesfully".format(scholardf.iloc[i,1]))
server.quit()
```
